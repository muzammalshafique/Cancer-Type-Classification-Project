## Exploratory Data Analysis:
Upon loading the datasets as pandas DataFrames, they were merged on the 'Samples' column, providing a unified DataFrame with gene expression features and cancer type labels.

### Balancing the Dataset:
Given the class imbalance observed, SMOTE (Synthetic Minority Over-sampling Technique) was applied to generate synthetic samples for minority classes, resulting in a balanced dataset.

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
import matplotlib.pyplot as plt

smote = SMOTE(sampling_strategy='not majority', k_neighbors=5)
X_res, y_res = smote.fit_resample(X, y)

# Then count and plot as before
counter_before = Counter(y)
counter_after = Counter(y_res)

df_smote = pd.DataFrame({
    'Class': [f'Class_{k}' for k in counter_before.keys()],
    'Samples before oversampling': counter_before.values(),
    'Samples after oversampling': counter_after.values()
})

# Set the index to 'Class' for proper plotting
df_smote.set_index('Class', inplace=True)

df_smote.plot(kind='bar')
plt.title('Class Distribution before and after SMOTE - MLP')
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
X_res_df = pd.DataFrame(X_res, columns=merged_df1.columns)
X_res_df['Target'] = y_res
merged_df = X_res_df
```

## Feature Selection:
Using mutual information, the top 500 features were selected based on their relevance to the target variable.

```python
from sklearn.feature_selection import mutual_info_classif

# Assuming df is your original DataFrame and 'FLAG' is the target variable
original_feature_names = merged_df.drop('Target', axis=1).columns.tolist()

info_gain = mutual_info_classif(X_res, y_res)
top_features_indices = info_gain.argsort()[-500:][::-1]  # Get indices of top 10 features
X_top_features = X_res[:, top_features_indices]

# Now retrieve the names of these top features
feature_names = [original_feature_names[i] for i in top_features_indices]
```

```python
merged_df = merged_df[feature_names]
merged_df['Output'] = y_res
merged_df
```

## Outlier Detection:
A threshold-based outlier detection was conducted, and outliers beyond 3 standard deviations from the mean were removed.

```python
Outliers = []
threshold = 3
m = len(merged_df.columns)-1
# Iterate through each column of the DataFrame merged_df
for i in range(m):
    in_outliers = []
    data = merged_df.iloc[:, i]
    mean = np.mean(data)
    std = np.std(data)
    data_array = merged_df.iloc[:, i].values
    # Iterate through each data point (j) in the column and its index (idx)
    for idx, j in enumerate(data_array):
        z_score = (j - mean) / std
        if np.abs(z_score) > threshold:
            in_outliers.append((idx, j))  # Append tuple of (index, value)
    Outliers.append(in_outliers)  # Append list of tuples for the column
```

```python
# Iterate through each row of the original dataset df
for i in range(len(Outliers)):
    outliers_indices = [t[0] for t in Outliers[i]]
    for j in outliers_indices:
        if j in merged_df.index:
            df = merged_df.drop(j)
        else:
            continue
```

## Machine Learning Model Training

```python
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

# Load your dataset
# merged_df = pd.read_csv('your_dataset.csv')

# Separate features and target
X = merged_df.drop('Output', axis=1)
y = merged_df['Output']

# Define a list of (name, classifier) tuples
classifiers = [
    ('Logistic Regression', LogisticRegression(max_iter=1000)),
    ('Random Forest', RandomForestClassifier()),
    ('Support Vector Machine', SVC()),
    ('Neural Network', MLPClassifier(max_iter=1000))
]

# Empty list to hold accuracy results
results = []

# Loop through classifiers, perform cross-validation, and store results
for name, clf in classifiers:
    # Create a pipeline that will scale the data before fitting the classifier
    pipeline = make_pipeline(StandardScaler(), clf)
    # Perform cross-validation and then take the mean of the scores
    cv_scores = cross_val_score(pipeline, X, y, cv=5, scoring='accuracy')
    results.append((name, cv_scores.mean()))

# Sort the results by the mean accuracy
sorted_results = sorted(results, key=lambda x: x[1], reverse=True)

# Print the results
for name, accuracy in sorted_results:
    print(f"{name}: {accuracy:.4f}")
```
